```python
df1_updated=pd.DataFrame(columns=['Duration','Rating'])
df1_updated['Duration']=df1['duration']
df1_updated['Rating']=df1['imdbRating']
df1_updated['dur']=np.log10(df1_updated['Duration'])
df1_updated['Rating'].fillna(0,inplace=True)
df1_updated['dur'].fillna(0,inplace=True)
df1_updated['Duration'].fillna(0,inplace=True)
```

```python
df2_updated=pd.DataFrame(columns=['Duration','Rating'])
df2_updated['Duration']=df2['duration']
df2_updated['Rating']=df2['imdb_score']
df2_updated['dur']=np.log10(df2_updated['Duration'])
df2_updated['Rating'].fillna(0,inplace=True)
df2_updated['dur'].fillna(0,inplace=True)
df2_updated['Duration'].fillna(0,inplace=True)
```

```python
kmeans1 = KMeans(n_clusters=3).fit(df1_updated)
y_kmeans1= kmeans1.predict(df1_updated)

kmeans2 = KMeans(n_clusters=3).fit(df2_updated)
y_kmeans2= kmeans2.predict(df2_updated)
```

```python

plt.scatter(df1_updated.iloc[:, 2], df1_updated.iloc[:, 1], c=y_kmeans1, s=10,cmap='viridis_r')
centers1 = kmeans1.cluster_centers_
plt.scatter(centers1[:, 2], centers1[:, 1], c='black')
plt.xlabel('Duration')
plt.ylabel('Rating')
plt.title("imdb_data")
```

```python
plt.scatter(df2_updated.iloc[:, 2], df2_updated.iloc[:, 1], c=y_kmeans2, s=10,cmap='viridis_r')
centers2 = kmeans2.cluster_centers_
plt.scatter(centers2[:, 2], centers2[:, 1], c='black')
plt.xlabel('Duration')
plt.ylabel('Rating')
plt.title("movie_metadata")
```
